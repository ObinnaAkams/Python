# Shuffle Game

```python
#Example 3: 

from random import shuffle

def shuffle_list(my_list):
    
    shuffle(my_list)
    
    return my_list

#game list from which the player will make a guess for the position of 'O'
my_list = ['', 'O', '']

#shuffles the list before the player makes a guess selection
shuffle_list(my_list)    
```

```python
#function to take the players guess

def player_guess():
    
    guess=''
    
 #This "while loop" makes sure that the right input is used and always return '"Please select
#from 0, 1, or 2' when ever number which are not 0,1, or 2 is inputted    
    while guess not in ['0', '1', '2']: 
        guess = input("Please select from 0, 1, or 2: ")
        
        
    return int(guess) #'int' converts string to integer. Rem that input always returns a string
```

```python
#function to combine the above two functions

def check_guess(my_list,guess):
    if my_list[guess] == 'O':
        print('Correct')
        
    else:
        print('Not correct')
        print(my_list) #shows the right position in the list after a failed attempt by the user
```

```python
#CALL ALL THE EARLIER FUNCTIONS IN AN ORDERED MANNER: Compactmentalize above fucntions by calling them
# Game list
my_list = ['', 'O', ''] #I think not supposed to be called

# Shuffled list
mixed_list = shuffle_list(my_list)  

# Player guess
users_guess = player_guess()

# Check guess
check_guess(mixed_list,users_guess)
```

```python
#1:

import math

class Line():
    
    def __init__(self,c1,c2):
        self.c1 = coordinate1
        self.c2 = coordinate2
    
    def distance(self):
        return math.sqrt((self.c2[0] - self.c1[0])**2 + (self.c2[1] - self.c1[1])**2)
    
    def slope(self):
        return (self.c2[1] - self.c1[1])/(self.c2[0] - self.c1[0])
    
```

```python
coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)
```

```python
li.distance()
```

```python
li.slope()
```

#### Problem 1
Fill in the Line class methods to accept coordinates as a pair of tuples and return the slope and distance of the line.

```python
class line:
    
    def __init__ (self, c1, c2):
        self.c1 = c1
        self.c2 = c2
        
        
    def distance(self):
        x1,y1 = self.c1
        x2,y2 = self.c2
        return ((x2-x1)**2 + (y2-y1)**2) ** 0.5
       
    
    def slope(self):
        x1,y1 = self.c1
        x2,y2 = self.c2
        return (y2-y1)/(x2-x1)
```

```python
c1 = (3,2)
c2 = (8,10)
```

```python
li = line(c1,c2)
```

```python
li.distance()
```

```python
li.slope()
```

```python

class Cylinder:
    pi = 3.14
    
    def __init__(self,height,radius):
        self.height = height
        self.radius = radius
        
    def volume(self):
        return pi*(self.radius)**2 *self.height
    
    def surface_area(self):
        return 2*pi*self.radius*(self.radius + self.height)
        
```

* owner
* balance

and two methods:

* deposit
* withdraw

As an added requirement, withdrawals may not exceed the available balance.

Instantiate your class, make several deposits and withdrawals, and test to make sure the account can't be overdrawn.

```python
class bankAcct:
    
    def __init__(self, owner, balance):
        self.owner =  owner
        self.balance = balance
        
    def __str__(self):
        return f"Account owner: {self.owner} \nAvailable balance: {self.balance}: "
        
    def deposit(self, dep_amt):
        self.balance += dep_amt 
        return 'Deposit accepted'
        
    def withdraw(self, with_amt):
        if self.balance >= with_amt:
            self.balance -= with_amt
            return 'Withdrawal successful'
            
            
        else:
            return 'Insufficient funds'
        
    
            
        
    
```
